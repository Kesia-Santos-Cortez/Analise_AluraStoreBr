### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
total_loja = sum(loja["Preço"])
total_loja2 = sum(loja2["Preço"])
total_loja3 = sum(loja3["Preço"])
total_loja4 = sum(loja4["Preço"])

totalFrete = sum(loja["Frete"])
totalFrete2 = sum(loja2["Frete"])
totalFrete3 = sum(loja3["Frete"])
totalFrete4 = sum(loja4["Frete"])

saldofrete = total_loja - totalFrete
saldofrete2 = total_loja2 - totalFrete2
saldofrete3 = total_loja3 - totalFrete3
saldofrete4 = total_loja4 - totalFrete4
```

```python
lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
faturamentos = [total_loja, total_loja2, total_loja3, total_loja4]
despesas = [totalFrete, totalFrete2, totalFrete3, totalFrete4]
saldofrete = [saldofrete, saldofrete2, saldofrete3, saldofrete4]
cores = ["#D4F6FF", "#C5D3E8", "#D0E8C5", "#B3C8CF"]
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


df = pd.DataFrame({
    'Loja': lojas,
    'Faturamento': faturamentos,
    'Despesas': despesas,
    'Saldo': saldofrete
})

sns.set_style("whitegrid")

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(25, 6))


sns.barplot(data=df, x='Loja', y='Faturamento', palette=cores, ax=ax1)
ax1.set_title('Faturamento por Lojas')
ax1.set_xlabel('Lojas')
ax1.set_ylabel('Total (R$)')
for p in ax1.patches:
    ax1.text(p.get_x() + p.get_width() / 2, p.get_height(),
             f'R$ {p.get_height():,.2f}', ha='center', va='bottom', fontsize=8)


sns.barplot(data=df, x='Loja', y='Despesas', palette=cores, ax=ax2)
ax2.set_title('Despesas por Lojas')
ax2.set_xlabel('Lojas')
ax2.set_ylabel('Total (R$)')
for p in ax2.patches:
    ax2.text(p.get_x() + p.get_width() / 2, p.get_height(),
             f'R$ {p.get_height():,.2f}', ha='center', va='bottom', fontsize=8)

sns.barplot(data=df, x='Loja', y='Saldo', palette=cores, ax=ax3)
ax3.set_title('Saldo por Lojas')
ax3.set_xlabel('Lojas')
ax3.set_ylabel('Total (R$)')
for p in ax3.patches:
    ax3.text(p.get_x() + p.get_width() / 2, p.get_height(),
             f'R$ {p.get_height():,.2f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria


```python
import pandas as pd
loja['Loja'] = 'Loja 1'
loja2['Loja'] = 'Loja 2'
loja3['Loja'] = 'Loja 3'
loja4['Loja'] = 'Loja 4'

vendas = pd.concat([loja, loja2, loja3, loja4], ignore_index=True)
```

```python
resultado = vendas.groupby(['Loja', 'Categoria do Produto']).size().reset_index(name='Quantidade')
resultado.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(13, 8))
sns.set_style("whitegrid")

ax = sns.barplot(data=resultado, x='Categoria do Produto', y='Quantidade', hue='Loja', palette=cores)

# Rótulos de dados de forma simplificada
for p in ax.patches:
    ax.text(p.get_x() + p.get_width() / 2, p.get_height(),
            f'{int(p.get_height())}', ha='center', va='bottom', fontsize=10)

plt.title('Vendas por Categoria e Loja')
plt.xlabel('Categoria')
plt.ylabel('Quantidade Vendida')
plt.legend(title='Lojas')
plt.tight_layout()
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
total_loja = sum(loja["Avaliação da compra"])
total_loja2 = sum(loja2["Avaliação da compra"])
total_loja3 = sum(loja3["Avaliação da compra"])
total_loja4 = sum(loja4["Avaliação da compra"])

qtdAvaliacao = len(loja)
qtdAvaliacao2 = len(loja2)
qtdAvaliacao3 = len(loja3)
qtdAvaliacao4 = len(loja4)

media = total_loja / qtdAvaliacao
media2 = total_loja2 / qtdAvaliacao2
media3 = total_loja3 / qtdAvaliacao3
media4 = total_loja4 / qtdAvaliacao4
```

```python
medias = [media, media2, media3, media4]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({
    'Loja': lojas,
    'Avaliações': medias
})

sns.set_style("whitegrid")
ax = sns.barplot(data=df, y='Loja', x='Avaliações', palette=cores)

# Adicionando rótulos de dados
for p in ax.patches:
    ax.text(p.get_width(), p.get_y() + p.get_height() / 2,
            f' {p.get_width():,.2f}', ha='left', va='center', fontsize=10)

# Títulos e rótulos
plt.title('Avaliação Média por Lojas')
plt.xlabel('Média Avaliações')
plt.ylabel('Lojas')
plt.tight_layout()
plt.show()
```

# 5. Frete Médio por Loja


```python
total_loja = sum(loja["Frete"])
total_loja2 = sum(loja2["Frete"])
total_loja3 = sum(loja3["Frete"])
total_loja4 = sum(loja4["Frete"])

qtdfretespagos = len(loja["Frete"])
qtdfretespagos2 = len(loja2["Frete"])
qtdfretespagos3 = len(loja3["Frete"])
qtdfretespagos4 = len(loja4["Frete"])

mediaFrete = total_loja / qtdfretespagos
mediaFrete2 = total_loja2 / qtdfretespagos2
mediaFrete3 = total_loja3 / qtdfretespagos3
mediaFrete4 = total_loja4 / qtdfretespagos4

mediasF = [mediaFrete, mediaFrete2, mediaFrete3, mediaFrete4]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Dados (mantendo seus dados)
df = pd.DataFrame({
    'Loja': lojas,
    'Frete Médio': mediasF
})

# Plotando o gráfico de linhas
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
ax = sns.lineplot(data=df, x='Loja', y='Frete Médio', marker='o', linewidth=2.5, markersize=8, color="#197392")

# Adicionando rótulos de dados
for i in range(len(df)):
    ax.text(i, df['Frete Médio'][i], f'R$ {df["Frete Médio"][i]:.2f}',
            va='bottom', ha='center', fontsize=10)

# Títulos e rótulos
plt.title('Frete Médio por Lojas')
plt.xlabel('Lojas')
plt.ylabel('Frete Médio (R$)')
plt.tight_layout()
plt.show()
```

#6. Produto Mais e Menos Vendido por Loja

```python
resultado = vendas.groupby(['Loja', 'Produto']).size().reset_index(name='Quantidade')
resultado.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np


nomes_lojas = []
produtos_menor = []
quantidades_menor = []
produtos_maior = []
quantidades_maior = []


for loja in resultado['Loja'].unique():
    dados_loja = resultado[resultado['Loja'] == loja]

    # Produto menos vendido
    produto_menor = dados_loja.loc[dados_loja['Quantidade'].idxmin()]
    produtos_menor.append(produto_menor['Produto'])
    quantidades_menor.append(produto_menor['Quantidade'])

    # Produto mais vendido
    produto_maior = dados_loja.loc[dados_loja['Quantidade'].idxmax()]
    produtos_maior.append(produto_maior['Produto'])
    quantidades_maior.append(produto_maior['Quantidade'])

    nomes_lojas.append(loja)

# Definir posição das barras
x = np.arange(len(nomes_lojas))  # posições no eixo x
largura = 0.35  # largura de cada barra


plt.figure(figsize=(14, 7))
barras_menor = plt.bar(x - largura/2, quantidades_menor, width=largura, label='Menos Vendido', color='salmon')
barras_maior = plt.bar(x + largura/2, quantidades_maior, width=largura, label='Mais Vendido', color='seagreen')

# Adicionar rótulos nas barras
for i in range(len(nomes_lojas)):
    # Menor vendido
    plt.text(x[i] - largura/2, quantidades_menor[i] + 0.5,
             f"{produtos_menor[i]}\n({quantidades_menor[i]})",
             ha='center', va='bottom', fontsize=8)

    # Maior vendido
    plt.text(x[i] + largura/2, quantidades_maior[i] + 0.5,
             f"{produtos_maior[i]}\n({quantidades_maior[i]})",
             ha='center', va='bottom', fontsize=8)


plt.title('Produto Mais e Menos Vendido por Loja')
plt.xlabel('Lojas')
plt.ylabel('Quantidade Vendida')
plt.xticks(x, nomes_lojas)
plt.ylim(0, max(quantidades_maior) + 10)
plt.legend()
plt.tight_layout()
plt.show()
```

#7. Ticket Médio por loja

```python
faturamento = vendas.groupby('Loja')['Preço'].sum()
numero_vendas = vendas.groupby('Loja').size()
ticket_medio = faturamento / numero_vendas

print(ticket_medio)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({
    'Loja': lojas,
    'Ticket Médio': ticket_medio
})

plt.figure(figsize=(8, 6))
sns.set_style("whitegrid")
ax = sns.lineplot(data=df, x='Loja', y='Ticket Médio', marker='o', linewidth=2.5, markersize=8, color='#FE7743')

# Adicionando rótulos de dados
for i in range(len(df)):
    ax.text(i, df['Ticket Médio'][i], f'R$ {df["Ticket Médio"][i]:.2f}',
            va='bottom', ha='center', fontsize=10)

# Títulos e rótulos
plt.title('Ticket Médio por Lojas')
plt.xlabel('Lojas')
plt.ylabel('Total (R$)')
plt.tight_layout()
plt.show()
```
